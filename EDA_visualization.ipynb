```python
# Step 2: Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Load TF-IDF matrix
df = pd.read_csv('tfidf_vectors.csv')

# Step 4: Show shape & head
print(" Shape of data:", df.shape)
print(df.head())

# Step 5: Check for any null values
print("\n‚ùì Missing values:")
print(df.isnull().sum())

# Step 6: Plot top 20 most frequent words across all bios
word_scores = df.sum().sort_values(ascending=False).head(20)

plt.figure(figsize=(12,6))
sns.barplot(x=word_scores.values, y=word_scores.index, palette='viridis')
plt.title('Top 20 Most Common Words in Bios (TF-IDF Scores)')
plt.xlabel('TF-IDF Total Score')
plt.ylabel('Words')
plt.tight_layout()
plt.show()

# Step 7: Optional ‚Äì Save this as an image
plt.savefig('top_words_plot.png')
```
