```python
from google.colab import files
uploaded = files.upload()
# Day 2 - Read and Clean Bios
import pandas as pd
import re

# Read the uploaded file into a pandas DataFrame
# Assuming the uploaded file is a CSV and there's only one file uploaded
file_name = next(iter(uploaded))
df = pd.read_csv(file_name)


print("ðŸ“„ Original Data:")
print(df.head())

# STEP 2: Clean function - removes symbols, punctuation, and lowercases
def clean_bio(text):
    text = str(text).lower()                          # Convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)               # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()          # Remove extra spaces
    return text

# STEP 3: Apply cleaning to each bio
df['cleaned_bio'] = df['bio'].apply(clean_bio)

# Show side-by-side original and cleaned bios
print("\nðŸ§¹ Cleaned Bios:")
print(df[['bio', 'cleaned_bio']].head())

# STEP 4: Save cleaned data (optional for now)
df.to_csv('cleaned_bios.csv',index=False)
```
